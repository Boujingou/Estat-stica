# Exerc√≠cio 01

Determine a equa√ß√£o da reta de regress√£o para os dados do produto interno bruto e emiss√£o de di√≥xido de carbono usados na aula passada (Exerc√≠cio 04 do notebook da aula 22).

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
# Dados
PIB = [1.7, 1.2, 2.5, 2.8, 3.6, 2.2, 0.8, 1.5, 2.4, 5.9]
CO2 = [552.6, 462.3, 475.4, 374.3, 748.5, 400.9, 253.0, 318.6, 496.8, 1180.6]
# Diagrama de dispers√£o
plt.scatter(PIB, CO2, c='black')
plt.xlabel('PIB [trilh√µes de d√≥lares]')
plt.ylabel('Emiss√µes de CO2\n[milh√µes de toneladas m√©tricas]')
plt.grid()
# Cria√ß√£o do objeto regress√£o linear 
regressao_CO2 = LinearRegression()
# C√°lculo da reta de regress√£o
regressao_CO2.fit(np.array(PIB).reshape(-1,1), np.array(CO2).reshape(-1,1))
# Exibi√ß√£o dos coeficientes da reta
inclinacao = regressao_CO2.coef_[0][0]
intercepto = regressao_CO2.intercept_[0]
print(f'Reta de regressao: {inclinacao:.3f}x + {intercepto:.3f}')
# Exibindo a reta de regress√£o
x = np.linspace(min(PIB), max(PIB), 50).reshape(-1,1)
y_previsto = regressao_CO2.predict(x)
plt.plot(x, y_previsto, c='red')
plt.scatter(np.array(PIB).mean(), np.array(CO2).mean(), c='blue')
```

# Exerc√≠cio 02

Determine a equa√ß√£o da reta de regress√£o para os dados do n√∫mero de anos em que os ex-alunos passaram fora da faculdade e suas contribui√ß√µes anuais usados na aula passada (Exerc√≠cio 01 do notebook da aula 22).

```python
import pandas as pd
# Dados
anos_fora = pd.Series([1, 10, 5, 15, 3, 24, 30])
contribuicao = pd.Series([12.5, 8.7, 14.6, 5.2, 9.9, 3.1, 2.7])
# Diagrama de dispers√£o
plt.scatter(anos_fora, contribuicao, c='green')
plt.xlabel('Anos fora da faculdade')
plt.ylabel('Contribui√ß√£o anual [milhares de US$]')
plt.grid()
# Cria√ß√£o do objeto regress√£o linear 
regressao_cont = LinearRegression()
# C√°lculo da reta de regress√£o
regressao_cont.fit(np.array(anos_fora).reshape(-1,1), np.array(contribuicao).reshape(-1,1))
# Exibi√ß√£o dos coeficientes da reta
inclinacao = regressao_cont.coef_[0][0]
intercepto = regressao_cont.intercept_[0]
print(f'Reta de regressao: {inclinacao:.3f}x + {intercepto:.3f}')
# Exibindo a reta de regress√£o
x = np.linspace(min(anos_fora), max(anos_fora), 50).reshape(-1,1)
y_previsto = regressao_cont.predict(x)
plt.plot(x, y_previsto, c='red')
plt.scatter(np.array(anos_fora).mean(), np.array(contribuicao).mean(), c='blue')
```

Os teores cal√≥rico *x* e de a√ß√∫car *y* (em gramas) de 11 cereais matinais com alto teor de fibras s√£o relacionados na tabela a seguir. (Fonte: *Consumer

Determine a reta de regress√£o que descreve a quantidade de a√ß√∫car do cereal em fun√ß√£o do teor cal√≥rico. A seguir, use essa reta para prever a quantidade de a√ß√∫car contida em um cereal matinal com teor cal√≥rico de:

(a) *x* = 90 calorias

(b) *x* = 150 calorias

(c) *x* = 175 calorias

(d) *x* = 208 calorias

```python
# Dados
lista_calorias = [140, 200, 160, 170, 170, 190, 190, 210, 190, 170, 160]
calorias = np.array(lista_calorias).reshape(-1,1)
lista_acucar = [6, 9, 6, 9, 10, 17, 13, 18, 19, 10, 10]
acucar = np.array(lista_acucar).reshape(-1,1)
# Diagrama de dispers√£o
plt.scatter(lista_calorias, lista_acucar, c='blue')
plt.xlabel('Teor cal√≥rico do cereal matinal [cal]')
plt.ylabel('Quantidade de a√ß√∫car [g]')
plt.grid()
# Cria√ß√£o do objeto regress√£o linear 
regressao_acucar = LinearRegression()
# C√°lculo da reta de regress√£o
regressao_acucar.fit(calorias,acucar)
# Exibi√ß√£o dos coeficientes da reta
inclinacao = regressao_acucar.coef_[0][0]
intercepto = regressao_acucar.intercept_[0]
print(f'Reta de regressao: {inclinacao:.3f}x + {intercepto:.3f}')
# Exibindo a reta de regress√£o
x = np.linspace(min(calorias), max(calorias), 50)
y_previsto = regressao_acucar.predict(x)
plt.plot(x, y_previsto, c='red')
# Previs√µes desejadas
calorias_desejadas = np.array([90, 150, 175, 208]).reshape(-1,1)
acucar_previsto = regressao_acucar.predict(calorias_desejadas)
print('*** Valores previstos ***')
for prev in range(len(acucar_previsto)):
  print(f'{calorias_desejadas[prev][0]} calorias --> {acucar_previsto[prev][0]:.2f} g de a√ß√∫car')
# Plotando as previs√µes
plt.scatter(calorias_desejadas,acucar_previsto,c='green')
plt.xticks(list(range(100,220,10)))
```
