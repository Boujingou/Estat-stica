# Exercício 04

Um economista quer determinar se existe relação linear entre o produto interno bruto (PIB, em trilhões de dólares) de países e as respectivas emissões de dióxido de carbono (CO2, em milhões de toneladas métricas). Os dados encontram-se nas variáveis *PIB* e *CO2*. Calcule o coeficiente de correlação e interprete o resultado no contexto dos dados.

```python
# Dados
PIB = [1.7, 1.2, 2.5, 2.8, 3.6, 2.2, 0.8, 1.5, 2.4, 5.9]
CO2 = [552.6, 462.3, 475.4, 374.3, 748.5, 400.9, 253.0, 318.6, 496.8, 1180.6]
# Diagrama de dispersão
plt.scatter(PIB, CO2, c='black')
plt.xlabel('PIB [trilhões de dólares]')
plt.ylabel('Emissões de CO2\n[milhões de toneladas métricas]')
plt.grid()
plt.show()

### Cálculo do coeficiente de correlação
print('*** Cálculo do coeficiente de correlação ***\n')

## 1º Método - Usando NumPy
print('--> 1º Método - Usando NumPy')
import numpy as np
# Matriz de correlação (contém os coeficeintes de correlação)
matriz_r = np.corrcoef(PIB, CO2)
print('Matriz de correlação:')
print(matriz_r)
print(f'Coeficiente de correlação r(PIB, CO2) = {matriz_r[0, 1]:.3f}\n')

## 2º Método - Usando SciPy
print('--> 2º Método - Usando SciPy')
import scipy.stats as st
# Tupla contendo o coeficiente de correlação (1º elemento) e
# o p-valor (2º elemento), usado para testes de hipótese
r_p_valor = st.pearsonr(PIB,CO2)
print('Tupla contendo o coeficiente de correlação e o p-valor:')
print(r_p_valor)
print(f'Coeficiente de correlação r(PIB, CO2) = {r_p_valor[0]:.3f}\n')
# Também é possível desencapsular o conteúdo da tupla
# r, p_valor = st.pearsonr(PIB, CO2)

## 3º Método - Usando Pandas Series
print('--> 3º Método - Usando Pandas Series')
# Dados no formato Pandas Series
PIBs = pd.Series(PIB)
CO2s = pd.Series(CO2)
# Coeficiente de correlação
r = PIBs.corr(CO2s,method='pearson')
print(f'Coeficiente de correlação r(PIB, CO2) = {r:.3f}\n')
# Também poderia ser usado:
# r = CO2s.corr(PIBs, method='pearson')

## 4º Método - Usando Pandas DataFrames
print('--> 4º Método - Pandas DataFrames')
dados = pd.DataFrame({'PIB':PIB, 'CO2': CO2})
display(dados)
# Matriz de correlação
print('Matriz de correlação:')
matriz_r_df = dados.corr(method='pearson')
display(matriz_r_df)
# Coeficiente de correlação
r = matriz_r_df.at['PIB', 'CO2']
print(f'Coeficiente de correlação r(PIB, CO2) = {r:.3f}')
```

# Exercício 07

No Exercício 04, empregamos 10 pares de dados para encontrar $r\approx 0,912$. Vamos testar a significância desse coeficiente de correlação, usando $\alpha = 0,05$.

```python
# Basta empregar o 2º Método para o cálculo do coeficiente de correlação
alpha = 0.05
# H0: rho = 0
# H1: rho != 0
r, p_valor = st.pearsonr(PIB,CO2)
if p_valor <= alpha:
  print(f'Hipótese nula rejeitada ao nível de significância de {alpha*100:.2f}%')
else:
  print(f'Hipótese nula NÃO rejeitada ao nível de significância de {alpha*100:.2f}%')
```
